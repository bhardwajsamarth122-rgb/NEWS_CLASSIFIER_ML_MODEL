```python
import pandas as pd
```

```python
df= pd.read_csv("../Data/Processed/news_category_processed.csv")
```

```python
df['short_description']=df['short_description'].fillna('')
```

```python
df['headline'] = df['headline'].fillna('')
```

```python
df.drop(['headline_len', 'short_description_len'], axis=1,inplace=True)
```

```python
df.head()
```

```python
import re
def cleaning(text):
    text = text.lower()
    text = re.sub(r'^a-z\s', '', text)
    return text
```

```python
df['headline'] = df['headline'].apply(cleaning)
df['short_description'] = df['short_description'].apply(cleaning)
df['category'] = df['category'].apply(cleaning)
```

```python
df['text'] = df['headline'] + " " + df['short_description']
```

```python
df.drop(['headline', 'short_description'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df=df.drop_duplicates()
```
